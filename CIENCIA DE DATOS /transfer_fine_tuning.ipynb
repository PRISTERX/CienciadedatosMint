```python
from tensorflow.keras.applications.resnet50 import ResNet50

# cargar el modelo base incluyendo el top
modelo_base = ResNet50(weights='imagenet', include_top=True)
```

```python
modelo_base.summary()
```

```python
nuevas_imagenes = [
"/content/drive/MyDrive/transfer /perro.jpg",
"/content/drive/MyDrive/transfer /bus.jpg",
"/content/drive/MyDrive/transfer /vaso.jpg"
]
```

```python
print(nuevas_imagenes)
```

```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input
import numpy as np

imagenes = []

for ruta in nuevas_imagenes:
    
    # cargar imagen y redimensionar a 224x224
    img = image.load_img(ruta, target_size=(224,224))
    
    # convertir a arreglo
    img_array = image.img_to_array(img)
    
    # expandir dimensión
    img_array = np.expand_dims(img_array, axis=0)
    
    # aplicar preprocess_input
    img_array = preprocess_input(img_array)
    
    # guardar en la lista
    imagenes.append(img_array)
```

```python
print(len(imagenes))
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import decode_predictions

for i, ruta in enumerate(nuevas_imagenes):

    # mostrar imagen
    img = image.load_img(ruta, target_size=(224,224))
    plt.imshow(img)
    plt.axis("off")
    plt.show()

    # hacer predicción
    pred = modelo_base.predict(imagenes[i])

    # mostrar top 3
    resultados = decode_predictions(pred, top=3)[0]

    print("Top 3 predicciones:")
    for clase in resultados:
        print(clase[1], ":", round(clase[2]*100,2), "%")

    print("\n----------------------\n")
```

```python
import tensorflow as tf
import os

# carpeta donde quieres guardar el dataset
save_dir = '/content/drive/MyDrive/Dog_cats'

# URL del dataset
URL = 'https://www.kaggle.com/api/v1/datasets/download/birajsth/cats-and-dogs-filtered'

# descargar y extraer
path_to_zip = tf.keras.utils.get_file(
    'cats_and_dogs.zip',
    origin=URL,
    extract=True,
    cache_dir=save_dir
)
```

```python
PATH = os.path.join(os.path.dirname(path_to_zip), 'cats_and_dogs_filtered')
```

```python
train_dir = os.path.join(PATH, 'train')
validation_dir = os.path.join(PATH, 'validation')
```
